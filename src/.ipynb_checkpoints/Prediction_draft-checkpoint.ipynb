```python
#Feature: Country (Converting it from string to integer)
country_dict={}
country_list=[]
country_list = list(data['country'])
```

```python
# Country
le = preprocessing.LabelEncoder()
le.fit(country_list)
country= list(le.classes_)
array_val=le.transform(country) 

for i in range(len(country)):    
    country_dict.update({country[i]:array_val[i]})

print("key\n",country_dict['Aruba'])
```

```python
#Updating the data with string feature Country's corresponding integer value .

data["country_number"] =''
for i in range(len(country_list)): 
    #print(test_dict[test_list[i]])   
    data.loc[data['country'] == country_list[i], ['country_number']] = country_dict[country_list[i]]
```
